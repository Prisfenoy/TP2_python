## Selección del modelo final

Una vez ajustados y evaluados los modelos de Regresión Logística y Random Forest, resulta necesario comparar su desempeño utilizando un mismo conjunto de métricas. Esta comparación permite identificar cuál de los dos modelos presenta una mejor capacidad predictiva sobre los datos de prueba y, en consecuencia, seleccionar el modelo más adecuado para la clasificación de la variable respuesta.

A continuación, se presenta una tabla comparativa con las principales métricas de evaluación obtenidas por ambos modelos. A partir de estos resultados, se seleccionará el modelo con mejor desempeño para volver a ajustarlo utilizando la totalidad de los datos disponibles en `temporada1.parquet`. Posteriormente, dicho modelo será empleado para generar predicciones sobre `temporada2.parquet`, que constituye el conjunto de datos destinado a la validación principal del trabajo.

```python
# Paquete necesario
import pandas as pd

# Se crea la tabla
comparacion = pd.DataFrame({
    "Métrica": [
        "Accuracy",
        "Sensibilidad",
        "Especificidad",
        "VPP",
        "F1-score",
        "AUC",
        "Log-Loss"
    ],
    "Regresión Logística": [
        0.665,  
        0.640,  
        0.689,  
        0.650,  
        0.645,  
        0.734,
        0.605  
    ],
    "Random Forest": [
        0.893,  
        0.909,  
        0.879,  
        0.871,  
        0.890, 
        0.956, 
        0.270   
    ]
})

comparacion
```

A partir de la tabla anterior, se observa que el modelo **Random Forest** presentó un mejor desempeño que la **Regresión Logística** en todas las métricas de evaluación consideradas. En particular, obtuvo mayores valores de Accuracy, Sensibilidad, Especificidad, Valor Predictivo Positivo (VPP), F1-score y AUC. Además, alcanzó un Log-Loss considerablemente menor, indicando una mejor calidad en las probabilidades predichas.

Estos resultados muestran que el modelo Random Forest presenta una mayor capacidad para clasificar correctamente los lanzamientos según se produzca o no un *swing*, así como una mejor capacidad discriminativa entre ambas clases. En consecuencia, se selecciona este modelo como modelo final para el presente trabajo. Posteriormente, será ajustado nuevamente utilizando la totalidad de los datos de `temporada1.parquet` y empleado para generar las predicciones sobre `temporada2.parquet`, correspondiente al conjunto de validación principal.
